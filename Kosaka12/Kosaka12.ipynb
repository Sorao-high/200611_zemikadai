```python
L_NA=[]
n=0
N=len(RT)
M=len(UNA)
for i in range(N):
    for j in range(M):
        if RT.iloc[i][0]<=UNA.iloc[j][0]<=RT.iloc[i][0]+1:
            L_NA.append(UNA.iloc[j][0])
        else:
            pass
        
L_NA
```

```python
L_VP=[]
n=0
N=len(RT)
M=len(UVP)
for i in range(N):
    for j in range(M):
        if RT.iloc[i][0]<=UVP.iloc[j][0]<=RT.iloc[i][0]+1:
            L_VP.append(UNA.iloc[j][0])
        else:
            pass
        
L_VP
```

```python
#RTの値から0.01秒間隔で1秒間, インターバル内にLVPがあれば1を, なければ0をisi_VPにappendしたいが, 
#変数が3つ必要でひとつループが多くなってしまう
isi_VP=[]
LVP=pd.Series(L_VP)
N=len(RT)
n=0.01
for i in range(N):
    for k in range(101):
        for j in range(len(LVP)):
            if RT.iloc[i][0]<=LVP.iloc[j]<RT.iloc[i][0]+n:
                isi_VP.append(1)
            else:
                isi_VP.append(0)
        n=0.01*k+0.02
print(isi_VP)
```
